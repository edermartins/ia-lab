## NLTK Tokenizer and Stemmer

Crie uma coluna no dataframe `news_2016` contendo os tokens para cada um dos textos. Os tokens devem estar representados pelo radical das palavras (stem).
Para tal, complete o conteúdo da função `tokenize`.

### Execução 9

```python
# Inicializei o stemmer RSLP e as stopwords fora da função para não recriar
# todas as vezes
import re
stemmer = RSLPStemmer()

def tokenize(text: str) -> List:
  """
  Função 'tokenize' utiliza `nltk.tokenize.word_tokenize`

  Returns:
    - Lista de tokens stemmed (`nltk.stem.RSLPStemmer`)
    IMPORTANTE: Somente tokens alfabéticos serão retornardos
  """
  #escreva seu código aqui
  text = text.lower()
  text = re.sub(r'[^a-z\s]', '', text)
  tokens = word_tokenize(text)
  custom_stopwords = set(stopwords.words('portuguese'))
  cleaned_tokens = [token for token in tokens if token not in custom_stopwords and len(token) > 2]
  stemmed_alpha_tokens = [stemmer.stem(token) for token in cleaned_tokens]

  return stemmed_alpha_tokens

news_2016.loc[:, 'nltk_tokens'] = news_2016.text.progress_map(tokenize)
```

## Realize a Lematização usando SPACY

O modelo NLP do spacy oferece a possiblidade de lematizar textos em português (o que não acontece com a biblioteca NLTK). Iremos criar uma lista de tokens
lematizados para cada texto do nosso dataset. Para tal, iremos retirar as
stopwords, usando uma função que junta stopwords provenientes do NLTK e do Spacy. Essa lista completa, é retornada pela função stopwords (e você não precisa mexer).

Já a função filter retorna True caso o token seja composto por caracters alfabéticos, não estiver dentro da lista de stopwords e o lemma resultante não estiver contido na lista `o", "em", "em o", "em a" e "ano"`.

Crie uma coluna chamada `spacy_lemma` para armazenar o resultado desse pré-processamento.

### Execução 12

```python
def stopwords() -> Set:
    """
    Retornar a lista completa de stopwords
    """
    return set(list(nltk.corpus.stopwords.words("portuguese")) + list(STOP_WORDS))

complete_stopwords = stopwords()

def filter(w: spacy.lang.pt.Portuguese) -> bool:
    """
    Filtra as stopwords e tokens não desejados

    Args:
        w (spacy.tokens.Token): Um token do spacy.

    Returns:
        bool: True se o token passar pelos filtros, False caso contrário.
    """
    # Critério 1: O token deve ser composto apenas por caracteres alfabéticos
    if not w.is_alpha:
        return False

    # Critério 2: O texto em minúsculas do token não deve ser uma stopword
    if w.text.lower() in complete_stopwords:
        return False

    # Critério 3: O lemma em minúsculas do token não deve estar na lista de lemas indesejados
    undesired_lemmas = ["o", "em", "em o", "em a", "ano"]
    if w.lemma_.lower() in undesired_lemmas:
         return False

    # Se passar por todos os critérios, o token é aceito
    return True


def lemma(doc: spacy.lang.pt.Portuguese) -> List[str]:
    """
    Aplica a Lemmatização nos tokens do texto

    Args:
        doc (spacy.tokens.doc.Doc): Um objeto Doc do spacy

    Returns:
        List[str]: Uma lista de lemmas padronizados (em minúsculas) dos tokens que passaram no filtro
    """
    lemmas_list = []
    for token in doc:
        # Filtra cada token
        if filter(token):
            # Se o token passar no filtro, adiciona seu lemma
            lemmas_list.append(token.lemma_.lower())

    return lemmas_list

# Função lemma para criar a coluna 'spacy_lemma'
news_2016.loc[:, 'spacy_lemma'] = news_2016.spacy_doc.progress_map(lemma)

# Exibindo as primeiras linhas do DataFrame com a nova coluna 'spacy_lemma'
print("DataFrame news_2016 com a nova coluna 'spacy_lemma':")
print(news_2016[['text', 'spacy_lemma']].head())

# Exibindo algumas linhas com a coluna 'spacy_lemma'
print("Coluna 'spacy_lemma' para algumas notícias:")
for i in range(min(5, len(news_2016))):
    print(f"Notícia {i}: {news_2016.iloc[i]['spacy_lemma']}")
```

## Reconhecimento de entidades nomeadas

Crie uma coluna `spacy_ner` que armazene todas as organizações (APENAS organizações) que estão contidas no texto.

### Execução 13

```python
def NER(doc: spacy.lang.pt.Portuguese):
  """
  Retorna a lista das organizações para o documento spacy
  Args:
      doc (spacy.tokens.doc.Doc): Um objeto Doc do spacy

  Returns:
      List[str]: Uma lista de strings, onde cada string é o texto de uma entidade do tipo "ORG"
  """
  organizations_list = []
  for ent in doc.ents:
      # Verifica se o rótulo da entidade é 'ORG'
      if ent.label_ == "ORG":
          organizations_list.append(ent.text)

  return organizations_list

news_2016.loc[:, 'spacy_ner'] = news_2016.spacy_doc.progress_map(NER)

# Mostrando as primeias linhas do DataFrame news_2016 com a nova coluna 'spacy_ner'
print("DataFrame news_2016 com a nova coluna 'spacy_ner':")
print(news_2016[['text', 'spacy_ner']].head())

# Exibir notícias que tenham pelo menos uma organização detectada
print("Mostrando algumas linhas com a coluna 'spacy_ner' (Organizações) para algumas notícias:")
news_with_orgs = news_2016[news_2016['spacy_ner'].apply(lambda x: len(x) > 0)].head()

if not news_with_orgs.empty:
    for index, row in news_with_orgs.iterrows():
          print(f"Notícia (trecho): {row['text'][:150]}...")
          print(f"Organizações detectadas: {row['spacy_ner']}")
          print("-" * 20)
else:
    print("Nenhuma organização ('ORG') encontrada nas primeiras notícias com NER.")
```

## Bag-of-Words

Crie uma coluna `tfidf` no dataframe `news_2016`. Use a coluna `spacy_lemma` como base para cálculo do TFIDF.
O número máximo de features que iremos considerar é 5000. E o token, tem que ter aparecido pelo menos 10 vezes (`min_df`) nos documentos.

### Execução 14

```python
class Vectorizer:
  def __init__(self, doc_tokens: List):
    self.doc_tokens = doc_tokens
    self.tfidf = None


  def vectorizer(self):
      """
      converte a lista de tokens para tfidf vector
      Retornar tfidf vector e também atribui ele para self.tfidf
      """
      # concatenando a lista de tokens de cada documento em uma única string
      documents_as_strings = [' '.join(tokens) for tokens in self.doc_tokens]

      # O TfidfVectorizer trata as strings como documentos
      tfidf_transformer = TfidfVectorizer(max_features=5000, min_df=10)

      # Calculando o vocabulário e os valores IDF para todo o dataset
      tfidf_transformer.fit(documents_as_strings)

      # Atribuindo a instância do vectorizer ao self.tfidf (já com o fit executado)
      self.tfidf = tfidf_transformer

  def __call__(self):
    if self.tfidf is None:
        self.vectorizer()
    return self.tfidf

# Criando uma lista 'doc_tokens'
doc_tokens = news_2016.spacy_lemma.values.tolist()
# Vectorizando a lista
vectorizer = Vectorizer(doc_tokens)

def tokens2tfidf(tokens):
    tokens = ' '.join(tokens)
    array =  vectorizer().transform([tokens]).toarray()[0]
    return array


# Aplica a função tokens2tfidf a cada lista de lemas na coluna 'spacy_lemma'
# e armazena o vetor TF-IDF resultante na nova coluna 'tfidf'
news_2016.loc[:, 'tfidf'] = news_2016.spacy_lemma.progress_map(tokens2tfidf)

# Exibindo as primeiras linhas do DataFrame com a nova coluna 'tfidf'
print("DataFrame news_2016 com a nova coluna 'tfidf':")
print(news_2016[['spacy_lemma', 'tfidf']].head())

# Mostrando a forma do array 'tfidf' da primeira notícia
print(f"Forma do vetor TF-IDF para a primeira notícia: {news_2016.iloc[0]['tfidf'].shape}")

# A forma deve ser (5000,) ou menos se min_df > 10 reduzir o vocabulário abaixo de 5000
# Mostrando os 10 primeiros e os 10 últimos
print(f"Primeiros 10 valores do vetor TF-IDF da primeira notícia: {news_2016.iloc[0]['tfidf'][:10]}")
print(f"Últimos 10 valores do vetor TF-IDF da primeira notícia: {news_2016.iloc[0]['tfidf'][-10:]}")
```

### Atribua a cada text, um (e apenas um) tópic.

Crie uma coluna `topic` onde o valor é exatamente o tópico que melhor caracteriza o documento de acordo com o algoritmo de LDA.

### Execução 17

```python
def get_topic(tfidf: np.array):
    """
    Get topic for a lda trained model
    Args:
        tfidf (np.ndarray): O vetor TF-IDF do documento (array 1D).

    Returns:
        int: O índice do tópico dominante (0 a N_TOKENS-1).
    """
    # O método transform do LDA espera uma entrada 2D (mesmo que seja apenas um documento).
    # Portanto, passi o vetor tfidf dentro de uma lista: [tfidf]
    # O resultado do transform é um array 2D (1, N_TOKENS), onde cada valor é a probabilidade do tópico.
    topic_distribution = lda.transform([tfidf])

    # np.argmax encontra o índice do valor máximo.
    # Como temos apenas um documento, topic_distribution[0] nos dá o array 1D de probabilidades de tópico para esse documento.
    dominant_topic_index = np.argmax(topic_distribution[0])

    return dominant_topic_index

news_2016['topic'] = news_2016.tfidf.progress_map(get_topic)

# Exiba as primeiras linhas do DataFrame com a nova coluna 'topic'
print("DataFrame news_2016 com a nova coluna 'topic':")
print(news_2016[['text', 'tfidf', 'topic']].head())

# Opcional: Contar a distribuição dos tópicos
print("Distribuição dos tópicos:")
print(news_2016['topic'].value_counts().sort_index())
```

## Crie uma nuvem de palavra para cada tópico.

Use as colunas `spacy_lemma` e  `topic` para essa tarefa.

### Execução 19

```python
def plot_wordcloud(text:str, ax:plt.Axes) -> plt.Axes:
    """
    Plot the wordcloud for the text/
    Arguments:
        -  text: string to be analised
        -  ax: plt subaxis
    Returns:
       plt.Axes: The modified matplotlib subaxis object.
    """
    if not text: # Verifica se o texto não está vazio
        return ax # Retorna o eixo sem plotar se não houver texto

    # Criando um objeto WordCloud
    wordcloud = WordCloud(
        background_color='white',
        width=800,
        height=600,
        max_words=100,
        contour_color='steelblue',
        prefer_horizontal=0.9
    ).generate(text)

    # Exibindo a nuvem de palavras
    ax.imshow(wordcloud, interpolation='bilinear')

    # Removendo os eixos para mostrar apenas a nuvem de palavras
    ax.axis("off")

    return ax

def plot_wordcloud_for_a_topic(topic:int, ax:plt.Axes) -> plt.Axes:
    topic_news = news_2016[news_2016['topic'] == topic]
    list_of_words = chain(*topic_news.spacy_lemma.values.tolist())
    string_complete = ' '.join(list_of_words)
    if not string_complete:
        return None
    return plot_wordcloud(string_complete, ax)

fig, axis = plt.subplots(3, 3, figsize=(16, 12))

axis_  = axis.flatten()
for idx, ax in enumerate(axis_):
    ax_ = plot_wordcloud_for_a_topic(idx + 1, ax)
    if ax_ is None:
        plt.delaxes(ax)
        continue
    ax.set_title(f"Tópico {idx + 1}")
fig.tight_layout()
```

```python
def plot_wordcloud_entities_for_a_topic(topic:int, ax:plt.Axes) -> plt.Axes:
    topic_news = news_2016[news_2016['topic'] == topic]
    list_of_docs = topic_news.spacy_ner.apply(lambda l : [w.replace(" ", "_") for w in l])
    list_of_words = chain(*list_of_docs)
    string_complete = ' '.join(list_of_words)
    if not len(string_complete):
        return None
    return plot_wordcloud(string_complete, ax)

fig, axis = plt.subplots(3, 3, figsize=(16, 12))

axis_  = axis.flatten()
for idx, ax in enumerate(axis_):
    ax_ = plot_wordcloud_entities_for_a_topic(idx + 1, ax)
    if ax_ is None:
        plt.delaxes(ax)
        continue
    ax.set_title(f"Tópico {idx + 1}")
fig.tight_layout()
```
